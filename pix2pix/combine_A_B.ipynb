```python
# iterative through different file tyeps

def get_data(ext, rootdir, subdir):
    '''
    This function returns a list:
    :param ext: list of 'jpg or whatever extension in use
    :param rootdir:  rootdir joining from
    :param subdir: subset directory or subdir
    :return: images list
    '''
    images = []
    for i in ext:
        # print(i)
#         print(os.path.join(rootdir, subdir, '*.'+i))
        alist = glob.glob(os.path.join(rootdir, subdir, '*.'+i))
#         print(len(alist))
        for fname in alist:
            if fname not in images:
                images.append(fname)
    return images
```
