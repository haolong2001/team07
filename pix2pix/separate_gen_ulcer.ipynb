```python
# resize all images from 256 by 256 to 128 by 128
# remove multiclass colouir by thresholding 

rootdir = os.path.join(os.getcwd(), 'results', 'footulcer_multiclass_256', 'big', 'strain')

masks_list = os.listdir(os.path.join(rootdir, 'masks'))
images_list = os.listdir(os.path.join(rootdir, 'images'))

destdir = os.path.join(os.getcwd(), 'results', 'footulcer_multiclass_256', 'big', 'train_big')
os.makedirs(destdir, exist_ok=True)
os.makedirs(os.path.join(destdir, 'images'), exist_ok=True)
os.makedirs(os.path.join(destdir, 'masks'), exist_ok=True)

for x in masks_list:
    mask = cv2.cvtColor(cv2.imread(os.path.join(rootdir, 'masks', x)), cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(mask, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    mask_resized = cv2.cvtColor(cv2.resize(thresh, (128, 128), interpolation=cv2.INTER_NEAREST_EXACT), cv2.COLOR_GRAY2RGB)
    plt.imsave(os.path.join(destdir, 'masks', x), mask_resized)
    
for y in images_list:
#         print(x)
    image = cv2.cvtColor(cv2.imread(os.path.join(rootdir, 'images', y)), cv2.COLOR_BGR2RGB)
    image_resized = cv2.resize(image, (128, 128), interpolation=cv2.INTER_NEAREST_EXACT)
    plt.imsave(os.path.join(destdir, 'images', y), image_resized)
        
        
        
```
